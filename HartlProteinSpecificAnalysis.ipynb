```python
path = '/home/jhdavis/hansen/data/2015_09_01-HartlAnalysis/'
peptides = pandas.read_csv(path+'peptides.txt', sep='\t', usecols = columns)
```

```python
def getRows(df, protein):
    toReturn = peptides[peptides['Leading razor protein']==protein].index.values
    if len(toReturn) == 0:
        toReturn = peptides[peptides['Gene names']==protein].index.values
    if len(toReturn) == 0:
        raise ValueError('Could not find any hits for protein or gene name ' + protein)
    else:
        return toReturn
```

```python
def getPLValues(df, rows):
    wtTPs = ['01','04','06','12']
    wtPL = []
    wtR = []
    wtPLTotal = []
    denom = [1, 0.134, 0.08839, 0.0625]
    
    for it, i in enumerate(wtTPs):
        wtPL.append((peptides.ix[rows]['Intensity H PulseLabeling.D'+i]/peptides.ix[rows]['Intensity L PulseLabeling.D'+i]).replace(
                    [np.inf, -np.inf], np.nan).dropna().values)
        try:
            wtR.append((peptides.ix[rows]['Intensity L WTday'+i]/peptides.ix[rows]['Intensity H WTday'+i]).replace(
                    [np.inf, -np.inf], np.nan).dropna().values)
        except:
            wtR.append([((numpy.median((peptides.ix[rows]['Intensity L WTday01']/peptides.ix[rows]['Intensity H WTday01']).replace(
                    [np.inf, -np.inf], np.nan).dropna().values)) + 
                        (numpy.median((peptides.ix[rows]['Intensity L WTday06']/peptides.ix[rows]['Intensity H WTday06']).replace(
                    [np.
                     inf, -np.inf], np.nan).dropna().values)))/2])
        
        wtPLTotal.append((peptides.ix[rows]['Intensity H PulseLabeling.D'+i]/peptides.ix[rows]['Intensity L PulseLabeling.D'+i].replace(
                    [np.inf, -np.inf], np.nan).dropna().values)/denom[it])
    

    return {'WTPL':[wtTPs, wtPL], 'WTR':[wtTPs, wtR], 'WTPLTotal':[wtTPs, wtPLTotal]}
```

```python
def plotPL(PLResults, geneOfInterest, yMax=None, xMax=14, colors=None, saveName=None):
    f = pylab.figure(figsize = (10,4))
    reps = [PLResults['WTPL'], PLResults['WTR'], PLResults['WTPLTotal']]
    repNames = ['turnover (H/L)', 'abundance', 'turnover/avgSynth']

    for rs, rep in enumerate(reps):
        a = f.add_subplot(1,3,rs+1)
        medY = []
        for it, x in enumerate(rep[0]):
            ys = rep[1][it]
            xs = [int(x)]*len(ys)
            medY.append(numpy.median(ys))
            a.scatter(xs, ys, alpha=0.75, color = colors[rs], s=15)
        a.plot(rep[0], medY, color = colors[rs])

        a.set_xlim(0, xMax)
        a.set_xlabel('days')
        if rs == 0:
            a.set_ylabel(geneOfInterest + ' value')
        if yMax != None:
            a.set_ylim(0, yMax)
        a.set_title(repNames[rs], color=colors[rs])
        vizLib.cleanAxis(a, ticksOnly=True)
        pylab.tight_layout()
        if saveName != None:
            pylab.savefig(saveName)
    return f
```

```python
def plotGenePL(peptides, geneOfInterest, yMax=None, colors=['black', 'red', 'blue'], saveName=None):
    try:
        rows = getRows(peptides, geneOfInterest)
    except ValueError:
        print 'Could not find data for gene/protein "' + geneOfInterest + '" this could result from "double-entries."'
        print 'Check the file "allGenes.txt" for more information.'
        return 
    repoResults = getPLValues(peptides, rows)
    plotPL(repoResults, geneOfInterest, yMax=yMax, xMax=14, colors=colors, saveName=saveName)
    
```

```python
def getRepoValues(df, rows):
    wtTPs = ['01','06','12']

    r1 = []
    r2 = []
    r3 = []
    r4 = []
    
    for i in wtTPs:
        r1.append((peptides.ix[rows]['Intensity L ReplicatesDay'+i+'-1']/peptides.ix[rows]['Intensity H ReplicatesDay'+i+'-1']).replace(
                    [np.inf, -np.inf], np.nan).dropna().values)
        r2.append((peptides.ix[rows]['Intensity L ReplicatesDay'+i+'-2']/peptides.ix[rows]['Intensity H ReplicatesDay'+i+'-2']).replace(
                    [np.inf, -np.inf], np.nan).dropna().values)
        r3.append((peptides.ix[rows]['Intensity L ReplicatesDay'+i+'-3']/peptides.ix[rows]['Intensity H ReplicatesDay'+i+'-3']).replace(
                    [np.inf, -np.inf], np.nan).dropna().values)
        r4.append((peptides.ix[rows]['Intensity L ReplicatesDay'+i+'-4']/peptides.ix[rows]['Intensity H ReplicatesDay'+i+'-4']).replace(
                    [np.inf, -np.inf], np.nan).dropna().values)

    return {'r1':[wtTPs, r1], 'r2':[wtTPs, r2], 'r3':[wtTPs, r3], 'r4':[wtTPs, r4]}
```

```python
def plotReproducibility(repoResults, geneOfInterest, yMax=None, xMax=30, colors=None, saveName=None):
    f = pylab.figure(figsize = (10,4))
    reps = [repoResults['r1'], repoResults['r2'], 
               repoResults['r3'], repoResults['r4']]
    repNames = ['rep1', 'rep2', 'rep3', 'rep4']

    for rs, rep in enumerate(reps):
        a = f.add_subplot(1,4,rs+1)
        medY = []
        for it, x in enumerate(rep[0]):
            ys = rep[1][it]
            xs = [int(x)]*len(ys)
            medY.append(numpy.median(ys))
            a.scatter(xs, ys, alpha=0.75, color = colors[rs], s=15)
        a.plot(rep[0], medY, color = colors[rs])

        a.set_xlim(0, xMax)
        a.set_xlabel('days')
        if rs == 0:
            a.set_ylabel(geneOfInterest + ' WT_abundance')
        if yMax != None:
            a.set_ylim(0, yMax)
        a.set_title(repNames[rs], color=colors[rs])
        vizLib.cleanAxis(a, ticksOnly=True)
        pylab.tight_layout()
        if saveName != None:
            pylab.savefig(saveName)
    return f
```

```python
def plotGeneReproducibility(peptides, geneOfInterest, yMax=None, colors=['black', 'red', 'blue', 'green'], saveName=None):
    try:
        rows = getRows(peptides, geneOfInterest)
    except ValueError:
        print 'Could not find data for gene/protein "' + geneOfInterest + '" this could result from "double-entries."'
        print 'Check the file "allGenes.txt" for more information.'
        return 
    repoResults = getRepoValues(peptides, rows)
    plotReproducibility(repoResults, geneOfInterest, yMax=yMax, xMax=14, colors=colors, saveName=saveName)
    
```

```python
def getTimeCourseValues(df, rows):
    wtTPs = ['01','06','12','17','22']
    daf2TPs = ['01','06','12','17','22','27']
    daf16TPs = ['01','06','12','17']
    hsf1TPs = ['01','06','12']

    wtR = []
    daf2R = []
    daf16R = []
    hsf1R = []
    
    for i in wtTPs:
        wtR.append((peptides.ix[rows]['Intensity L WTday'+i]/peptides.ix[rows]['Intensity H WTday'+i]).replace(
                    [np.inf, -np.inf], np.nan).dropna().values)
    
    for i in daf2TPs:
        daf2R.append((peptides.ix[rows]['Intensity L DAF2day'+i]/peptides.ix[rows]['Intensity H DAF2day'+i]).replace(
                    [np.inf, -np.inf], np.nan).dropna().values)

    for i in daf16TPs:
        daf16R.append((peptides.ix[rows]['Intensity L DAF16day'+i]/peptides.ix[rows]['Intensity H DAF16day'+i]).replace(
                    [np.inf, -np.inf], np.nan).dropna().values)

    for i in hsf1TPs:
        hsf1R.append((peptides.ix[rows]['Intensity L HSF1day'+i]/peptides.ix[rows]['Intensity H HSF1day'+i]).replace(
                    [np.inf, -np.inf], np.nan).dropna().values)

    return {'WT':[wtTPs, wtR], 'DAF-2':[daf2TPs, daf2R], 'DAF-16':[daf16TPs, daf16R], 'HSF-1':[hsf1TPs, hsf1R]}
```

```python
def getTimeCourseValuesRel(df, rows):
    wtTPs = ['01','06','12','17','22']
    daf2TPs = ['01','06','12','17','22','27']
    daf16TPs = ['01','06','12','17']
    hsf1TPs = ['01','06','12']

    wtR = []
    daf2R = []
    daf16R = []
    hsf1R = []
    
    for i in wtTPs:
        wtR.append((numpy.log2((peptides.ix[rows]['Intensity L WTday'+i]/peptides.ix[rows]['Intensity H WTday'+i])/
                   (peptides.ix[rows]['Intensity L WTday01']/peptides.ix[rows]['Intensity H WTday01']))).replace(
                    [np.inf, -np.inf], np.nan).dropna().values)
    
    for i in daf2TPs:
        daf2R.append((peptides.ix[rows]['Intensity L DAF2day'+i]/peptides.ix[rows]['Intensity H DAF2day'+i]).replace(
                    [np.inf, -np.inf], np.nan).dropna().values)

    for i in daf16TPs:
        daf16R.append((peptides.ix[rows]['Intensity L DAF16day'+i]/peptides.ix[rows]['Intensity H DAF16day'+i]).replace(
                    [np.inf, -np.inf], np.nan).dropna().values)

    for i in hsf1TPs:
        hsf1R.append((peptides.ix[rows]['Intensity L HSF1day'+i]/peptides.ix[rows]['Intensity H HSF1day'+i]).replace(
                    [np.inf, -np.inf], np.nan).dropna().values)

    return {'WT':[wtTPs, wtR], 'DAF-2':[daf2TPs, daf2R], 'DAF-16':[daf16TPs, daf16R], 'HSF-1':[hsf1TPs, hsf1R]}
```

```python
def plotTimeCourse(timeCourseResults, geneOfInterest, yMax=None, xMax=30, colors=None, saveName=None):
    f = pylab.figure(figsize = (10,4))
    mutants = [timeCourseResults['WT'], timeCourseResults['DAF-2'], 
               timeCourseResults['DAF-16'], timeCourseResults['HSF-1']]
    mutNames = ['WT', 'DAF-2', 'DAF-16', 'HSF-1']

    for ms, mut in enumerate(mutants):
        a = f.add_subplot(1,4,ms+1)
        medY = []
        for it, x in enumerate(mut[0]):
            ys = mut[1][it]
            xs = [int(x)]*len(ys)
            medY.append(numpy.median(ys))
            a.scatter(xs, ys, alpha=0.75, color = colors[ms], s=15)
        a.plot(mut[0], medY, color = colors[ms])

        a.set_xlim(0, xMax)
        a.set_xlabel('days')
        if ms == 0:
            a.set_ylabel(geneOfInterest + ' abundance')
        if yMax != None:
            a.set_ylim(0, yMax)
        a.set_title(mutNames[ms], color=colors[ms])
        vizLib.cleanAxis(a, ticksOnly=True)
        pylab.tight_layout()
        if saveName != None:
            pylab.savefig(saveName)
    return f
```

```python
def plotGeneAbundance(peptides, geneOfInterest, yMax=None, colors=['black', 'red', 'blue', 'green'], saveName=None):
    try:
        rows = getRows(peptides, geneOfInterest)
    except ValueError:
        print 'Could not find data for gene/protein "' + geneOfInterest + '" this could result from "double-entries."'
        print 'Check the file "allGenes.txt" for more information.'
        return 
    timeCourseResults = getTimeCourseValues(peptides, rows)
    plotTimeCourse(timeCourseResults, geneOfInterest, yMax=yMax, xMax=30, colors=colors, saveName=saveName)
    
```
